```python
import pandas as pd
import numpy as np
```

```python
dataset2 = [['Milk','Eggs','Cool Drinks'],['Milk','Bread','Eggs','Jam'],['Milk', 'Cool Drinks','Bread','Eggs'],['Eggs','Cool Drinks','Wheat bread','Jam']]
```

```python
dataset2
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
myarray = te.fit(dataset2).transform(dataset2)
```

```python
df = pd.DataFrame(myarray,columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
Features_Items = apriori(df,min_support=0.1,use_colnames=True)
```

```python
Features_Items
```

```python
Features_Items['length'] = Features_Items['itemsets'].apply(lambda x:len(x))
```

```python
Features_Items
```

```python
Features_Items[(Features_Items['support'] >=0.50) & (Features_Items['length'] >2) ]
```

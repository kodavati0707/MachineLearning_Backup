```python
import pandas as pd
import numpy as np
```

```python
dataset = [['Milk','Bread','Wheat Bread'],['Jam','Bread','Butter'],['Milk','Jam','Bread','Butter'],['Jam','Butter']]
```

```python
dataset
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
te_array = te.fit(dataset).transform(dataset)
```

```python
df= pd.DataFrame(te_array,columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequency_items= apriori(df,min_support=0.1,use_colnames=True)
```

```python
frequency_items
```

```python
frequency_items['length'] = frequency_items['itemsets'].apply(lambda x:len(x))
```

```python
frequency_items
```

```python
frequency_items[(frequency_items['length']>=3) & (frequency_items['support']>=0.20)]
```

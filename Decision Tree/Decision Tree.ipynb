```python
import pandas as pd
import numpy as np
```

```python
Titanic_data= pd.read_csv('train.csv')
```

```python
Titanic_data.head()
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
gender = label_encoder.fit_transform(Titanic_data['Sex'])
```

```python
gender
```

```python
Titanic_data['Sex'] = gender
```

```python
Titanic_data.head()
```

```python
from sklearn import tree
```

```python
model_tree = tree.DecisionTreeClassifier()
```

```python
Titanic_data.head()
```

```python
model_tree.fit(X = pd.DataFrame(gender), y =Titanic_data['Survived'])
```

```python
with open('Dtree.dot','w') as f:
    f = tree.export_graphviz(model_tree,feature_names=["Sex"],out_file=f);
```

```python
check = pd.DataFrame([Titanic_data['Sex'], Titanic_data['Pclass']])
check.head()
```

```python
preditors = pd.DataFrame([gender, Titanic_data['Pclass']]).T
```

```python
preditors
```

```python
model_tree.fit(X = preditors, y =Titanic_data['Survived'])
```

```python
with open('Dtree_2.dot','w') as f:
    f = tree.export_graphviz(model_tree,feature_names=["Sex","Pclass"],out_file=f);
```
